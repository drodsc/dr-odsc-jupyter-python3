### Building a Predictive Model Factory

#### Definition (Gartner is free to Quote me here!)

A model factory, in the context of data science, is a system, or a set of procedures that automatically generates predictive models with little or no human intervention. Model factories can have multiple layers of complexity which we can call modules. One module might be training models while other modules could be deploying or retraining the models.

### Visualize the series

```python
for department in df['Department'].unique()[0:4]:
    fig, ax = pyplot.subplots(figsize=(10,10))
    sns.lineplot(x='Date',y='Sales',data=df.loc[df['Department'] == department],hue='Department')
```

### Cluster Together the different Series

```python
mean_per_department = df.groupby('Department')['Sales'].mean()
std_per_department = df.groupby('Department')['Sales'].std()

def normalise(row):
    return (row['Sales'] - mean_per_department[row['Department']])/std_per_department[row['Department']]
```

```python
df['Sales_normalised'] = df.apply(lambda row: normalise(row),axis=1)
X = df[['Date','Department','Sales_normalised']].pivot(index='Department',columns='Date',values='Sales_normalised').fillna(0)

# calculate the correlation matrix
corr = X.transpose().corr()
fig, ax = pyplot.subplots(figsize=(20,20))
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True)
```

### Which department belongs to each cluster?

```python
import scipy.cluster.hierarchy as spc
pdist =  spc.distance.pdist(corr)
linkage = spc.linkage(pdist, method='complete')
idx = spc.fcluster(linkage, 0.5 * pdist.max(), 'distance')

cluster_dictionary = {k: [] for k in range(1,idx.max()+1)}

#Create dicitonary with clusters
for i,j in zip(df['Department'].unique(), idx):
    cluster_dictionary[j].append(i)
```

### Visualize each cluster of series

```python
for cluster in cluster_dictionary.values():
    fig, ax = pyplot.subplots(figsize=(10,10))
    sns.lineplot(x='Date',y='Sales',data=df.loc[df['Department'].isin(cluster)],hue='Department')
```

### At last, let us create the model factory!

```python
for cluster in cluster_dictionary.values():
    project_name = 'Series_{}'.format(cluster)
    data = df.loc[df['Department'].isin(cluster)]
    
    proj = dr.Project.create(sourcedata=data,
                             project_name = project_name)
    
    print('Project ID: {}'.format(proj.id))
    
    #Set partitioning
    time_partition = dr.DatetimePartitioningSpecification(
        datetime_partition_column='Date',
        multiseries_id_columns=['Department'],
        use_time_series=True,
        feature_derivation_window_start = -30, #Could hyperparametrize
        feature_derivation_window_end= 0, #Could hyperparametrize
        forecast_window_start = 1,
        forecast_window_end = 10,
        number_of_backtests=3,
        windows_basis_unit='ROW'
        )
    
    #Initiate project
    proj.set_target(
        target='Sales',
        partitioning_method=time_partition,
        max_wait=4200,
        worker_count=-1)
```
